```python
# Series Indexing:-  Series indexing in pandas is simply the way you access or select data from a Series using labels or positions

#Positive Indexing
x = pd.Series([12, 13, 18, 17, 14, 16, 15, 19, 20])
print("\nPositive Integer Indexing:- ", x[1])
print("\nPositive Integer Indexing:- ", x[8])

#Negative Indexing:- Series always works on Positive indexing not negative indexing

# print("\nNegative Indexing:- ", x[-1]) :- Give Error
 
print("\n Marks Datasets:- \n", marks)
print("\n Negative Indexing in marks datasets:- \n", marks.iloc[-2])


 
print("\nKohli Datasets:- \n", kohli)
print("\n Negative Indexing in kohli datasets:- \n", kohli.iloc[-2])

```

```python
"""Slicing in pandas means extracting a subset of data (range of values) from a Series or DataFrame—based on positions or labels.

Think of it as:
👉 “Mujhe pura data nahi, sirf ek range chahiye.”"""

print("\nLast 5 Runs of Kohli:- \n",kohli[-5:])
print("\nLast 10 Movies in data sets:- \n",blwd[-10:])


"""Fancy indexing in pandas is a power move to select multiple specific elements using a list (or array) of indices or labels—instead of a continuous range like slicing."""

print("\nRun made in 1, 2, 3, 4, 5, 6, 7 matches by kohli:- \n",kohli.iloc[[1, 3, 4, 5, 6, 7]])


```

```python
# Editing Series:- 

#using Indexing

print("\nRename Subject name:- \n")
marks = marks.rename({1 : 'English'})
print(marks)

print("\n Editing series based on label of data:- \n")
marks.loc['maths'] = 89
print(marks)

print("\n Editing serieds based on index of data:- \n")
marks.iloc[2] = 55
print(marks)
# What if an index does not exist:- then it create ones!!
marks['sst'] = 74
print("\n", marks)

# Editing multiple item using slicing:- 

kohli.iloc[2:4, 0] = [100,100] 
print("\nRun of matches number 2 to 3 are:- \n", kohli)

kohli.iloc[2:5, 0] = [0,0,0]
print("\nUpdate runs of matches 2 to 4\n", kohli)

#fancy indexing:- 
print(len(kohli.iloc[2:4]))

kohli[[0,4,5]] = [0,0,0]
print("\nupdated run 0:- \n",kohli)
kohli
```
