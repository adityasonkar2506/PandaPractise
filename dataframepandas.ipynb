```python
def trackrecord(ipl_matches, team1, team2):
    
    matches = ipl_matches[
        ((ipl_matches['Team1'] == team1) & (ipl_matches['Team2'] == team2)) |
        ((ipl_matches['Team1'] == team2) & (ipl_matches['Team2'] == team1)) 
    ]

    total = matches.shape[0]

    team1_wins = matches[matches['WinningTeam'] == team1].shape[0]
    team2_wins = matches[matches['WinningTeam'] == team2].shape[0]

    return {
        'Total Matches': total,
        team1: team1_wins,
        team2: team2_wins
    }

result = trackrecord(ipl_matches, 'Chennai Super Kings', 'Mumbai Indians')
print(result)
```
