Mathematical Methods

```python
#sum
movie.sum() 
# this is not logical answer 
```

```python
print(stud)
print("\n")
print(stud.sum()) # this apply sum to all the columns
print("\nSum apply on rows:- \n", stud.sum(axis=1)) # axis = 1 pefrom mathematical operation on rows


# mean():- 
print("\nMean:- \n", stud.mean()) # col wise
print("\nMean:- \n", stud.mean(axis=1)) # row wise

#min():- 
print("\nMinimum value column wise:- \n", stud.min())
print("\nMinimum value Row Wise:- \n", stud.min(axis=1))


#max():-
print("\nMaximum value column wise:- \n", stud.max()) # col wise
print("\nMaximum value row wise:- \n", stud.max(axis=1))# row wise


#median():- It gives the middle value after sorting the data 

print("\nMedain value column wise:- \n", stud.median())
print("\nMedian value row wise:- \n", stud.median(axis=1))

#std():- standard deviatation(It measures how spread out your data is from the average.)

print("\nStandard Deviation Column Wise:- \n", stud.std()) #col wise
print("\nStandard Deviation Row wise:-\n", stud.std(axis=1)) #row wise

#var():- It measures how much the data varies (spread) — just like std, but in squared form.

print("\nVariance of data column wise:- \n", stud.var()) #col wise
print("\nVariance of data row wise:- \n", stud.var(axis=1)) # row wise

```
