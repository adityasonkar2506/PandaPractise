PANDAS 
Herer we are learning Pandas Series:- It is like a column in a table It is a 1-D array holding data of any type

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt

print("All libraries are installed successfully")
```

Series from lists:-


```python
# String:- 

country = ['India', 'Pakistan', 'China', 'Nepal', 'Myanmaar', 'Bangladesh']
print("String Dataype:- \n")
pd.Series(country)

```

```python
# Int Dataype

runs = [13, 24, 57, 84, 100, 140, 167, 189, 202]
print("Integer Dataype:- \n")
pd.Series(runs)
```

```python
# Custom Index:- 
marks = [65, 57, 89, 100]
subjects = ['maths', 'chemistry', 'physics', 'english']
print("Subject as index and marks as value!!\n")
pd.Series(marks, index=subjects)
```

```python
print("Assigning name variable:- name = Marks of Anushka!!\n")
marks = pd.Series(marks, index=subjects, name = 'Marks of Anushka')
marks
```

Series from Dictonary :- 

```python
print("Printing Series from Dictonary!!\n")
marks = {
    'maths': 65,
    'chemistry': 57,
    'physics': 89,
    'english': 100
}

marks_series = pd.Series(marks, name='Marks of Anushka!!')
marks_series
```

Series Attributes:- 

```python
# Size :- This attribute tell about size of series
print("Size of Series!!")
print(marks_series.size)

# dType:- This attribute tell about the datatype.
print("\nData Type of Series!!")
print(marks_series.dtype)

# Name:- This attribute tell about the name of series.
print("\nName of Series!!")
print(marks_series.name)

# is_unique:- This attribute tell about the uniqueness of series
print("\nUnique or not if yes:- True else False!!")
print(marks_series.is_unique)

print("\nNot Unique:-")
print(pd.Series([1,1,5,8,3,3]).is_unique)

#index:-  This Return all value of index
print("\nIndex return all values of index in series:- ")
print(marks_series.index)

#values:- This Returns all values of value
print("\nValue return all values of sereies!!")
print(marks_series.values)
print("\nType of this result is:- ", type(marks_series.values))
```

Series Methods:-

```python
# head and tail :- this attribute is use to give preview of data
# head :-  Returns the first few rows(by default 5 rows) of series
subs = pd.read_csv('subs.csv').squeeze()
kohli = pd.read_csv('kohli_ipl.csv').squeeze()
blwd = pd.read_csv('bollywood.csv').squeeze()

print("\n Subscriber Info:-")
print("\n",subs.head()) # by default it fetchs top 5 rows
print("\n",subs.head(20)) # by choice we can enter number of rows want to fetch


# Tail:- Return the last few rows(by default 5 rows) of series
print ("\nNumber of Subscriber in last 5 days:- ")
print("\n", subs.tail()) # by default it fetch last 5 rows 

print("\nNumber of Subscriber in last 20 days:-\n", subs.tail(20)) # last 20 days subscriber:- 


# sample:- randomly selects some rows/data from a Series by default always give one dataset from series and after reloding it frequently changes the dataset

print("\n Sample data set from Movies:- \n", blwd.sample())

# you can assign number of values by your choice 

print("\n Sample data set from Movies:- \n", blwd.sample(5))

# value_counts:-  counts how many times each values appears!!

#1st method:- 
result = pd.read_csv('bollywood.csv')['lead'].value_counts()
print("\nNumber of movies done by Actors:- \n",result)

#2nd mehthod:-
print("\n1Number of movies done by Actors:- \n", blwd['lead'].value_counts())


# sort_values :-  arrange values according to data

#Ascending Order
print("\nNumber of Run made by kohli in entire IPL from lowest to highest:- \n", kohli['runs'].sort_values())

#Descending Order 

print("\nNumber of Run made by kohli in entire IPL from highest to lowest:- \n", kohli['runs'].sort_values(ascending=False))

# print highest run made by kohli

print("\nHighest Run made by kohli in entire match:- \n", kohli['runs'].sort_values(ascending=False).head(1).values) 


# sort_values --> inplace --> runs
"""inplace=True original data ko modify karta hai, jabki
 inplace=False new sorted copy return karta hai."""

print("\n Original values:- \n", kohli['runs'].sort_values(ascending=False))

#using inplace to perform permanent changes:- 

print("\n After using inplace orginal data is modified:- \n", kohli['runs'].sort_values(inplace=True))
print(kohli)

# sort_index --> inplace --> movies

print("\n")
print("\nOrignal data of movie in sorted manner:- \n", blwd.sort_index())



```

Series Mathematical Methods:- 

```python
# count :- Total number of items in series excluding missing values!!
# size :- Total number of items in series including missing values!!

print("Total number of values in kohli datasets:- \n", kohli.count())

# sum 

print("\nNumber of subscribers gain in last one year:- \n", subs.sum())

# product

print("\n Product of Subscriber in last one year:- \n", subs.product())
"""The above product method give zero because maybe somedays i gain 0 subscriber.
multiplication of any number with 0 comes out to be zero!!"""


# Mean = (sum of all values) / (number of values)
print("\nMean(Average) of Subscribers:- \n","Subscriber gain on daily basis:- ", subs.mean())

# Median =  median() function is used to find the middle value of a dataset after arranging the data in order.
print("\nMedian of Virat Kohli Scores:- \n", kohli['runs'].median())

# Mode = mode() function is used to find the most frequently occurring value(s) in a dataset.

print("\nThe Actors who make the highest number of in dat filmsa set:- \n", blwd['lead'].mode())

# std(standard deviation):- std() function is used to calculate the standard deviation, which measures the spread of data around the mean.
print("\nThe standard deviation subscribers:-   \n", subs.std())

# var(variance) :- var() function is used to calculate the variance, which measures how far data values are spread from the mean.
print("\nThe variance of run of Virat Kohli:- \n", kohli['runs'].var())

# min(Minimum)
print("\nThe minimum number of subscriber gain in one day in last year:- \n", subs.min())

# max(Max)
print("\nThe maximum number of subscriber gain in one year in last year:- \n", subs.max())

# describe():- describe() function is used to generate a summary of statistical measures of a dataset.

print("\nDetails summary report of Kohli Runs:- \n", kohli['runs'].describe())
print("\nDetails summary report of subscriber:- \n", subs.describe())
```
