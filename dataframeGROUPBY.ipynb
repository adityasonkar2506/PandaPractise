GroupBy Object in Python:- 

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
movies = pd.read_csv('imdb-top-1000.csv')
movies.head()
```

```python
movies.shape
```

```python
# group movies on the basis of genre 
genres = movies.groupby('Genre')
```

```python
# Applying builtin aggregation function on groupby objects

genres.sum() 
```

```python
# min

genres.min()

"""by seeing the below data we can say ki Action Genre me sbse pehele release hui movie 1914 ki hai 
* Action Genre me sbse kam rating waali movie 7.6 imdb ki movie hai
* Drama me sbse pehel release hui movie 1917 ki hai
* Drama me sbse km runtime waali movie 64 minute ki hai 

"""
```

```python
# max()

genres.max()

"""
* Action me sbse latest releas hui movie 2019 ki hai
* Action me sbse High rating waali movie 9.0 ki hai
* Action me sbse high number of vote waali movie 2303232 ki hai
"""
```

```python
# method 1 to solve this

# find  the top 3 genres by total earning
print("Data is Grouped + Sum, not sorted\n")
print(movies.groupby('Genre').sum()['Gross']) # Unsorted


# we have to sort on the basis of values:- 
print("\nData is Grouped + Sum + Sorted from highest to lowest\n")
print(movies.groupby('Genre').sum()['Gross'].sort_values(ascending=False))

# top 3
print("\nTop 3 Genre with most earnings:- \n")
print(movies.groupby('Genre').sum()['Gross'].sort_values(ascending=False).head(3))



#Output:- Drama, Action, Comedy
```

```python
# method 2:- 

movies.groupby('Genre')['Gross'].sum().sort_values(ascending=False)
```

```python
# find the Genre with highest avg IMDB rating:- 
print("List of genre with highest avg IMDB rating:- \n")
print(movies.groupby('Genre')['IMDB_Rating'].mean().sort_values(ascending=False))

print("\nDisplay The one Genre with highest avg IMDB rating:- \n")
print(movies.groupby('Genre')['IMDB_Rating'].mean().sort_values(ascending=False).head(1))




```

```python
"""
Aggregation function:-  when you have to perform single calculation at a time
Multi Aggregation function:-  when you have to perform multiple calculation at a time

"""
genres.sum()
# this will give total sum of every columns in each group which make data useless 


```

```python
# agg()
# apply agg() using dict
# this is best way if we're applying custom aggregation 
# column1 pe kuch or , c2 pe kuch or, c3 pe kuch or

genres.agg(
    {
        'Runtime': 'mean',
        'IMDB_Rating': 'mean',
        'No_of_Votes': 'sum',
        'Gross': 'sum',
        'Metascore': 'min'
    }
)
```

```python
# agg() apply in list
#method:- 1
# genres.agg(['min', 'max', 'mean']) if this will give you error 

# use another
# if selecting multiple column then use double square brackets:-  '[[]]'
genres[['Runtime', 'IMDB_Rating', 'No_of_Votes', 'Gross', 'Metascore']].agg(['min', 'max', 'mean'])
```

```python
genres[['Runtime', 'IMDB_Rating', 'No_of_Votes', 'Gross']].agg(['min', 'max', 'mean'])
```

```python
# if single column ho use single squar bracket '[]'

genres['Runtime'].agg(['min', 'max', 'mean'])
```

```python
# Adding both syntax
#  merge syntax:- 

genres.agg(
    {
        'Runtime': ['mean', 'min', 'max'],
        'IMDB_Rating': 'mean',
        'No_of_Votes': ['sum', 'min', 'max', 'mean'],
        'Gross': 'sum',
        'Metascore': ['min', 'max', 'mean', 'sum']
    }
)
```

```python
# looping on groups:- Data ko pehle groups me divide karna using groupby(),
# phir har group ke upar ek-ek karke loop chalana.



data = {
    'Team' : ['India', 'Australia', 'India', 'Australia'],
    'Players' : ['Virat', 'Smith', 'Rohit', 'Warner'],
    'Runs' : [80, 95, 72, 45]
}

df = pd.DataFrame(data)

```

```python
df
```

```python
# loop
for team, group in df.groupby('Team'):
    print(team)
    print(group)
```

```python
# genre 

for group, data in genres:
    print(group)
    print(data)
```

```python
# some previous question can be solved using loop

# q:-  Highest rated movie of each genre





for group, data in genres:
    print(data[data['IMDB_Rating'] == data['IMDB_Rating'].max()])
    

```

```python
# split (apply) combine 
# apply -> builtin function


genres.apply(min)
```

```python
# find number of movies starting with A for each group
# method 
# split + apply
def fun(group):
    return group['Series_Title'].str.startswith('A').sum() 


    
```

```python
# combine = split + apply
genres.apply(fun)
```

```python
# print number of movies start with ('A')
# method2 using loop

for group, data in genres:
    print("Genre:", group)     
    print(data['Series_Title'].str.startswith('A').sum())
    print()
```

```python
#  find ranking of each movie in the group to IMDB score
# split + apply
def rank_movie(group):
    group['genre_rank'] = group['IMDB_Rating'].rank(ascending = False)
    return group
```

```python
# combine
genres.apply(rank_movie)
```

```python
#find normalized IMDB rating group wise
# split + apply
def normal(group):
    group['norm_rating'] = (group['IMDB_Rating'] - group['IMDB_Rating'].min()) / (group['IMDB_Rating'].max() - group['IMDB_Rating'].min())
    return group
#combine
genres.apply(normal)
```

```python
# practise on real datasets 
ipl = pd.read_csv('deliveries.csv')
ipl.head(5




)

```

```python
ipl.shape
```

```python
# 1. find the top 10 batsman in terms of runs
ipl.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False).head(10)
```

```python
# find the batsman with max no. of sixes

six = ipl[ipl['batsman_runs'] == 6]
six.groupby('batsman')['batsman'].count().sort_values(ascending=False).head()
```

```python
# find the batsman with most of 4's and 6's in last 5 overs
#method1
# wromg

fors = ipl[ipl['batsman_runs'] == 4]
fors.groupby('batsman')['batsman'].count().sort_values(ascending=False)

fors_six = fors.groupby('batsman')['batsman'].count().sort_values(ascending=False) + six.groupby('batsman')['batsman'].count().sort_values(ascending=False)
fors_six.tail(5).sort_values(ascending=False)

```

```python
# method2
death_overs = ipl[ipl['over'] > 15]

fours = death_overs[death_overs['batsman_runs'] == 4]
sixes = death_overs[death_overs['batsman_runs'] == 6]

four_sixes = (
    fours.groupby('batsman')['batsman'].count()+ 
    sixes.groupby('batsman')['batsman'].count()
)

four_sixes.sort_values(ascending=False).head()
```

```python
# method3
temp_df = ipl[ipl['over'] > 15]
temp_df[(temp_df['batsman_runs'] ==4) | (temp_df['batsman_runs'] == 6)]
temp_df.groupby('batsman')['batsman'].count().sort_values(ascending=False)
```
