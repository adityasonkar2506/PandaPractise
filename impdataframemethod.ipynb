```python
"""default behaviour of dropna:-
ki agar kisi bhi row me ek bhi missing value (Nan) hai to wo eniter row ko remove 
krdeta hai jisse problem ye hai ki bohot sara data remove hojaayga

but what we want is ki dropna() khaali wo row remove kre jis row ki saari value
missing (Nan) ho naki wo jaha ek ya do missing value ho baaki me values ho

"""

students

```

```python
# this define:- remove entire row if any missing value exist in particular row
print(students.dropna(how='any'))
print("\n")
# this define:-  remove entire row only if all value is Nan in row
print(students.dropna(how='all'))

```

```python
# in marks dataset as above table show that there is one row that is duplicated row 3 and 4 
# use in dataframe
marks.drop_duplicates()
# aftet applying drop_duplicates() one row is removed 
```

```python
# use in series

marks = pd.Series([85, 95, 77, 77, 95],
            ['Rahul', 'Arjun', 'Anjali', 'Anjali', 'Smith'])
print(marks)

print("\nRemoving duplicate row in series:- \n", marks.drop_duplicates())
```

```python
# drop_duplicates(key = 'first')
# Duplicate me first value rakho, baaki delete krdo 

print(marks)
print("\nKeeping first value of duplicates just remove last one:- \n", marks.drop_duplicates(keep='first'))

# drop_duplicates(key='last')
# Duplicate me last value rakho, baaki remove krdo

print("\n",marks)
print("\nKeeping last value of duplicates just remove first one:-  \n", marks.drop_duplicates(keep='last'))
```
